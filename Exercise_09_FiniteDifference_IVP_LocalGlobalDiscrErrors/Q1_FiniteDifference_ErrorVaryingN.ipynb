```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return np.exp(-x**2)

def df_exact(x):
    return -2 * x * np.exp(-x*x)
```

```python
def fd_central(f, z, h):
    return ( f(z+h) - f(z-h) ) / (2*h)

def fd_bwd(f, z, h):
    return ( f(z) - f(z-h) ) / h

def fd_fwd(f, z, h):
    return( f(z+h) - f(z) ) / h
```

# Error for varying N

## Central Finite Difference

```python
N = 1 # initial N
k = 10
error_list = np.zeros(k)
N_list     = np.zeros(k)
decrease_order4 = np.zeros(k)
decrease_order2 = np.zeros(k)


for i in range(k):
    h = 2 / N
    x = np.arange(-1, 1, h)
    # compute the exact and the numerical first derivative
    D_exact = [df_exact(z) for z in x]
    D_num = [fd_central(func, z, h) for z in x]
    error_list[i] = np.sqrt(sum([(D_exact[j] - D_num[j])**2 for j in range(N)])) / N
    if (i>0):
        decrease_order4[i] = decrease_order4[i-1] / 4
        decrease_order2[i] = decrease_order2[i-1] / 2

    else:
        decrease_order4[0] = error_list[0]
        decrease_order2[0] = error_list[0]


    N_list[i] = N
    N = N*2 # iteratively double N in each loop
    


plt.plot(np.log2(N_list), error_list, color="red")
plt.plot(np.log2(N_list), decrease_order4, "--", color="black")
plt.plot(np.log2(N_list), decrease_order2, "--", color="blue")

plt.legend(["error central FD", "convergence order O(h^2)", "convergence order O(h)"])
None

Ecentral = error_list
```

# Second Derivative

```python
def second_derivative_fd_central(f, z, h):
    return ( f(z+h) - 2 * f(z) + f(z-h) ) / (h**2)
```

```python
def ddf_exact(x):
    return np.exp(-x**2) * (4*x**2 - 2)
```

```python
N = 1 # initial N
k = 10
error_list = np.zeros(k)
N_list     = np.zeros(k)
decrease_order4 = np.zeros(k)
decrease_order2 = np.zeros(k)


for i in range(k):
    h = 2 / N
    x = np.arange(-1, 1, h)
    # compute the exact and the numerical second derivative
    D_exact = [ddf_exact(z) for z in x]
    D_num = [second_derivative_fd_central(func, z, h) for z in x]
    error_list[i] = np.sqrt(sum([(D_exact[j] - D_num[j])**2 for j in range(N)])) / N
    if (i>0):
        decrease_order4[i] = decrease_order4[i-1] / 4
        decrease_order2[i] = decrease_order2[i-1] / 2

    else:
        decrease_order4[0] = error_list[0]
        decrease_order2[0] = error_list[0]


    N_list[i] = N
    N = N*2 # iteratively double N in each loop
    


plt.plot(np.log2(N_list), error_list, color="red")
plt.plot(np.log2(N_list), decrease_order4, "--", color="black")
plt.plot(np.log2(N_list), decrease_order2, "--", color="blue")

plt.legend(["error second derivative central FD", "convergence order O(h^2)", "convergence order O(h)"])

None

Ecentral = error_list
```
