```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initialize the arrays
N = 300
x1 = np.zeros(N)
x2 = np.zeros(N)

# ICs
x1[0] = 1
x2[0] = 0
```

```python
h = 0.1 # time step

# Solve equation for damped harmonic oscillator
for j in range(N-1):
    x1[j+1] = x1[j] + h * x2[j]
    x2[j+1] = (1-0.5*h) * x2[j] - h * x1[j] 
```

```python
t = np.arange(0, N*h, h)
```

```python
plt.plot(t, x1)
plt.xlabel("time")
plt.ylabel("space x")
None
```
