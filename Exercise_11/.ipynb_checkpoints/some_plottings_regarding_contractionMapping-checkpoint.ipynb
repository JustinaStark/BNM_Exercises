```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(x):
    return x + np.sin(x)
```

```python
a, b = 1, 5
x = np.arange(a, b, 0.1)
y = [g(x[i]) for i in range(len(x))]
F = [g(x[i]) - x[i] for i in range(len(x))]
```

```python
plt.plot(x, y)
plt.plot(x, x)
plt.plot(x, F)
labels = ["g(x)", "y=x", "F(x) = g(x) - x"]
plt.legend(labels)
# plt.plot(x, np.cos(x))
plt.title('g(x) is a contraction map in the interval [a, b] \nand has therefore one unique fixed point in I at F(c) = 0. \nc has to be within the interval I according to the mean value theorem.')
None
```
