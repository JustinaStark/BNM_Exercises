```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
A = np.array([[3, 2], [2, 6]])
b = np.array([2, -8])
c = 0
```

```python
def cost_fct(A, b, c, x):
    f = 0.5 * x.T @ A @ x - x.T @ b + c
    return f
```

```python
x1 = np.linspace(-4, 6)
x2 = np.linspace(4, -6)

x = np.array([x1, x2]).T
```

```python
x.shape
```

```python
F = np.zeros((len(x), len(x)))
for i in range(len(x1)):
    for j in range(len(x2)):
        x = np.array([x1[i], x2[j]])
        F[i, j] = cost_fct(A, b, c, x)
```

```python
F.shape
```

```python
minimum = np.linalg.solve(A, b)
```

```python
plt.contour(x1, x2, F)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Contour plot of F(x)')
plt.plot(minimum[0], minimum[1], '*')
None
```

```python
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X, Y = np.meshgrid(x1, x2)


# Plot the surface.
surf = ax.plot_surface(X, Y, F, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# # Customize the z axis.
# ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
ax.set_title('Surface plot of cost function F')

# # Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
A = np.array([[3, 2, 1], [2, 3, 2], [1, 2, 3]])
```

```python
# A symmetric matrix is positiv definite iff all its eigenvalues are > 0
l = np.linalg.eigvals(A)
print("Is A positive definite? ", all(l>0))
print("The eigenvalues of A are ", l)
```

```python
eigvals = np.linalg.eigvals((A.T @ A))
eigvals
```

```python
kappa = np.sqrt(max(eigvals))/np.sqrt(min(eigvals))
kappa
```

```python
np.sqrt(kappa)
```

```python
alpha = (np.sqrt(kappa) - 1.0) / (np.sqrt(kappa) + 1.0)
```

```python
np.log10(10**(-10) / 2) / np.log10(alpha)
```
